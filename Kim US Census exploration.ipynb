```python
import numpy as np
import pandas as pd

n = np.arange(1,11)
files = []

for each in n:
    files.append(f'us states{each}.csv')
    
data = [pd.read_csv(file) for file in files]

for i in range(len(data)):
    to_drop = data[i].columns[data[i].columns.str.contains("Note")]
    data[i].drop(columns=to_drop, inplace=True) #remove columns we don't need
    data[i] = data[i].transpose() #pivot so features are the columns
    data[i].dropna(how="any", axis=1, inplace=True)
    
data_combined = pd.concat(data)
data_combined.columns = data_combined.iloc[0]
data_combined.drop("Fact", axis=0, inplace=True)
data_combined.drop("United States", axis=0, inplace=True)
data_combined.to_csv('census_quickfacts.csv')
```
